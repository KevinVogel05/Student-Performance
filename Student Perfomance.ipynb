## Procedure

```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pickle
import sklearn
from sklearn import linear_model
from matplotlib import style

#data
df = pd.read_csv('input/student-mat.csv', sep=";")
```

```python
df.head()
```

```python
display(df.columns, df.dtypes, df.shape, df.describe())
```

```python
#change sex to binary (0=Female, 1=Male)
df['sex'].replace(('M', 'F'), (1, 0), inplace=True)
```

```python
#choose relevant data
data = df[['G1','G2','G3','studytime','failures','absences']]
data2 = df[['G1','G2','G3','studytime','failures','absences','age','health','sex','traveltime']]

#target
predict = 'G3'

#split data
X = np.array(data.drop([predict], 1))
y = np.array(data[predict])

X2 = np.array(data2.drop([predict], 1))
y2 = np.array(data2[predict])
```

```python
#generate best modell
def generate_best_modell(X, y, data):
    best = 0
    for _ in range(500):
        #train the modell
        X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.1)
        linear = linear_model.LinearRegression()
        linear.fit(X_train, y_train)
        acc = linear.score(X_test, y_test)

        if acc > best:
            best = acc
            print(best)
            name = str(best)            
            #save modell
            with open('models/' + data + 'lr' + name[2:6] + '.pickle', 'wb') as f:
                pickle.dump(linear, f)

##generate_best_modell(X, y, '1')

#load modell
pickle_in = open('models/2lr9732.pickle', 'rb')
linear = pickle.load(pickle_in)
```
